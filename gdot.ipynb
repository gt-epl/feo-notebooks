```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../gdot/site1.csv')
df2 = pd.read_csv('../gdot/site2.csv')
df3 = pd.read_csv('../gdot/site3.csv')
df4 = pd.read_csv('../gdot/site4.csv')
```

```python
fig, ax = plt.subplots()
df1['Mon'].plot(ax=ax)
df2['Mon'].plot(ax=ax)
df3['Mon'].plot(ax=ax)
df4['Mon'].plot(ax=ax)
ax.set_xticks(range(24))
ax.set_xlabel('hours')
ax.set_ylabel('# Cars')
ax.legend([f'site_{i}' for i in range(4)])
```
